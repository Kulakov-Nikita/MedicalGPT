```python
import urllib
import requests

f = {"q":"кожа","type":"clinical", "offset":"0", "limit":"10"}
url = "https://demo.onco-reg.ru/docs/api?" + urllib.parse.urlencode(f)

respond = requests.get(url)
answer = respond.json()
number_of_pdf = answer['result']['total']
documents = answer['result']['documents']
#print(number_of_pdf)
links = [doc['link'] for doc in documents]
```

```python
PDFs = []

for link in links:
    PDFs.append("\n".join([page.extract_text() for page in PdfReader(io.BytesIO(requests.get(link).content)).pages]))
```
